```python
## no libraries allowed except for numpy and biopython 
## no Bio.pairwise2 allowed

from Bio.Align import substitution_matrices
from numpy import zeros

def globalAlignment(x, y, sub, indel):
    """ Calculate the global alignment between sequences x and y using
        dynamic programming using the substitution matrix <sub>, and <indel> penalty
        Return the full dynamic programming table. """
    
    
    D = zeros((len(x)+1, len(y)+1), dtype=int) # initialize a numpy matrix with zeroes
 



    # No need for backtracking matrix, just check 3 incoming edges when backtracking

    # Start at the first row and populate the max score

        # f i r s t w o r d 
    #   s
    #   e 
    #   c
    #   o
    #   n
    #   d



    # Populate first row

    for j in range(1,len(y) + 1):
        previousNodeValue = D[0][j-1]


        D[0][j] = previousNodeValue + indel
    
    # Populate first column

    for i in range(1,len(x) + 1):

        previousNodeValue = D[i-1][0]
        D[i][0] = previousNodeValue + indel


    # Populate everything else besides first row and first column

    for i in range(1,len(x) + 1):

        for j in range(1,len(y) + 1):

            leftNodeValue = D[i][j-1]
            upNodeValue = D[i-1][j]
            diagonalNodeValue = D[i-1][j-1]

            leftTotalValue = leftNodeValue + indel
            upTotalValue = upNodeValue + indel

            letter1 = x[i-1]
            letter2 = y[j-1]


            # Everytime you do a diagonal move, you have to fetch it from the "sub" matrix

            diagonalEdgeValue = sub[letter1][letter2]

            diagonalTotalValue = diagonalNodeValue + diagonalEdgeValue

            D[i][j] = max(max(leftTotalValue,upTotalValue),diagonalTotalValue)


    return D




def tracebackGlobal(V, x, y, sub, indel):
    """ Trace back the alignment from position (len(x), len(y)) in global alignment matrix V """
    align1, align2 = '', ''
    
    
    
    i, j = len(x), len(y) # start from the bottom right cell
    
    # print(i,j)


    while i > 0 and j > 0:

        # Check each neighbor

        upNodeValue = V[i-1][j] + indel
        leftNodeValue = V[i][j-1] + indel

        letter1 = x[i-1]
        letter2 = y[j-1]

        diagonalNodeValue = V[i-1][j-1] + sub[letter1][letter2]

        # print(diagonalNodeValue)


            
            

        if upNodeValue >= diagonalNodeValue and upNodeValue >= leftNodeValue:

            align1 += letter1
            align2 += "-"

            i -= 1


        elif leftNodeValue >= diagonalNodeValue and leftNodeValue >= upNodeValue:
            align1 += "-"
            align2 += letter2

            j -= 1
        else:

            align1 += letter1
            align2 += letter2

            i -=1 
            j -=1

            

    while j > 0:

        letter2 = y[j-1]
        align1 += "-"
        align2 += letter2
        j-=1

    while i > 0:
        letter1 = x[i-1]
        align1 += letter1
        align2 += "-"
        i-=1



    return align1[::-1]+'\n'+align2[::-1]  # reverse the strings











f = open("hw4_test1.txt", "r")                                # open the file
blosum62 = substitution_matrices.load('BLOSUM62') # possible matrices are ['BENNER22', 'BENNER6', 'BENNER74', 'BLOSUM45', 'BLOSUM50', 'BLOSUM62', 'BLOSUM80', 'BLOSUM90', 'DAYHOFF', 'FENG', 'GENETIC', 'GONNET1992', 'HOXD70', 'JOHNSON', 'JONES', 'LEVIN', 'MCLACHLAN', 'MDM78', 'NUC.4.4', 'PAM250', 'PAM30', 'PAM70', 'RAO', 'RISLER', 'SCHNEIDER', 'STR', 'TRANS']
# blosum62 is a numpy matrix
# here is how to access the substitution cost for the aminoacid pair (A,M)
# print(blosum62.get('A').get('M'))
p1 = f.readline().strip()                                     # read the first string
p2 = f.readline().strip()                                     # read the second string

M = globalAlignment(p1, p2, blosum62, -5)                     # compute the dynamic programming matrix betwen p1 and p2, using the substitution matrix and indel penalty
# The dynamic programming matrix for this input should be
#[[  0  -5 -10 -15 -20 -25 -30]
# [ -5  -2  -6 -11 -16 -21 -26]
# [-10  -3  -5  -7 -12 -12 -17]
# [-15  -8   2  -3  -7 -12 -14]
# [-20 -13  -3   6   1  -4  -9]
# [-25 -18  -8   1   7   2  -3]
# [-30 -23 -13  -4   2   6   1]
# [-35 -28 -18  -9   2   1   4]
# [-40 -33 -23 -14  -3   1  -1]
# [-45 -38 -28 -19  -8   1   0]
# [-50 -43 -33 -24 -13  -4   8]]
# print(M)
print(M[len(p1),len(p2)])                                     # alignment score in on lower right corner cell
print(tracebackGlobal(M, p1, p2, blosum62, -5))               # print alignment
```
