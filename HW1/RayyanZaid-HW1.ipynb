<br>
How do you explain the fact that this gene is fragmented in 10 pieces along the X chromosome (i.e., the BLAST alignment is broken into 10 ranges)?

## YOUR ANSWER HERE

This is because of splicing. In eukaryotic cells, genes have exons and introns.  The initial RNA transcript (pre-mRNA) includes both exons and introns. During splicing, the introns are removed, and exons are joined together to form the mature mRNA. This mature mRNA, which contains only exons, is what gets translated into protein. When this mRNA is aligned back to the DNA using BLAST, it looks fragmented because the introns in the DNA are not present in the mature mRNA. This results in the BLAST alignment showing discontinuous or broken ranges along the chromosome, corresponding to the exonic regions.
